**3 Use transpose function :**

```python
# create array and use transpose function
# transpose funct use for change colum into row and row into column

arr2 = np.array([[1,2,3,4,8],[1,6,8,9,7,]])
arr2.transpose()
```

**4 create array and combine it using hstack func ()**

```python
# create two arratys and cobine it
# use hstack func()
# it will write in just one line
arr3 = np.array([1,2,3,4,8])
arr4 = np.array([1,6,8,9,7,])

# two methods for using hstack function

np.hstack([arr3, arr4])
          # OR
# arr3.hstack(arr4)
```

```python
arr5= np.array([1,2,3,4,8])
arr6 = np.array([1,6,8,9,7,])
combine=np.hstack([arr5, arr6])
arr5_new, arr6_new = np.hsplit(combine,[len(arr5)])
```
