**Index , Slicing , iteration**





**1 Create array :**

```python
# Create array in arange method
import numpy as np
arr1 = np.arange(24)
arr1
```

**2 use ,reshape method ():**

```python
# use reshape method
# we make row and columns :
arr2 = arr1.reshape(6,4)
arr2
```

**find numbers through index and slicing:**

```python
# give index two which is in row wise
# arr2[2]         # output is [8,9,10,11]
# arr2[:2]        # output is 1st 2 rows
# arr2[:,2]         # find third column : output is [2,6,10,14,18,22]
# arr2[:,1:3]         find index 1 colum and index 2 colum: out put is 1 and 2 index
```

**Iteration :**

```python
# create array with exisiting
# iteration

arr3 = arr2
arr3
```

```python
#  iteration in array using for loop with np.nditer()
#  it print numbers in asending order
for i in np.nditer(arr3):
       print(i)
```
