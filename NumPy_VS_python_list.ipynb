**1  Create list and numpy Array and check how much storage they take **

```python
import numpy as np
import sys         # import system time
# import time        # import time for find the time of execution of programme that how much time take python list and numpy array

lista = range(100)

arr1 = np.arange(100,dtype=np.int32)

print (sys.getsizeof(87)*len(lista))     # print for lista and python list store  more memory

print(arr1.itemsize*arr1.size)           # print for arr1 and numpy array store less memory
```
