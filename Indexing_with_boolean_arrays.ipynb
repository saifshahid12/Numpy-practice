**Create array of random number :**

```python
import numpy as np
# genearte randome number array
# 1 to 100 with 20 items

arr1 = np.random.randint(low =1 , high = 100 , size =20).reshape(4,5)
arr1
```

**get row or colums using index :**

```python
# get row and number of row from array

arr1[0][3]       # 0 is index of row and 3 is index of 3rd number in 0 row
```

**Conditional indexing :**

```python
# this is boolean method
# use condition if greater or less then number
# out put will be true and false
arr1>50
```

**Overlaping in existing :**

```python
# overlaping in existig code
# chek if there is trues false
# show true values in outpu
# using verlaping method

arr1[arr1>50]
```

**select great number form existing code:**

```python
# select great numbers which great then 50 and must be odd

arr1[(arr1>50) & (arr1%2!=0)] # modulus formula use for odd number
```

**make zero number in the array:**

```python
# make number 0 which is greter then 50 and is odd
# take from existing code

arr1[(arr1>50)&(arr1%2!=0)]=0
arr1
```
